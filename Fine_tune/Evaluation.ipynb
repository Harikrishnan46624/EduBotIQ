```python
from bert_score import score
```

```python
generated_text = ["RNNs are a type of recurrent neural network, where each layer learns to remember the output of the previous layer. Feedforward neural networks, on the other hand, learn to predict the next input based on the current output. RNNs are commonly used in sequence modeling tasks, such as language modeling and time series prediction."]
orginal_text = ["RNNs have connections that create cycles in the network, allowing them to process sequential data, whereas feedforward networks process data without considering the order."]
```

```python
P, R, F1 = score(generated_text, orginal_text, lang='en', verbose=True)
```

```python
print(f"Precision: {P.mean()}")
```

```python
print(f"Recall: {R.mean()}")
```

```python
print(f"F1: {F1.mean()}")
```

```python
generated_text = ["Natural Language Processing (NLP) is the study of how computers understand and interpret human language. It involves techniques like machine learning, text mining, and natural language generation. NLP is essential for applications like chatbots, machine translation, and natural language processing."]
orginal_text = ["Natural Language Processing is a field of artificial intelligence that focuses on the interaction between computers and human languages, enabling machines to understand, interpret, and generate human-like text."]
```

```python
P, R, F1 = score(generated_text, orginal_text, lang='en', verbose=True)
```

```python
print(f"Precision: {P.mean()}")
print(f"Recall: {R.mean()}")
print(f"F1: {F1.mean()}")
```
